# Spectogram : Specto Frequency : Amp vs Frequency

```python
plt.figure(figsize=(20, 5))
plt.title('Specshow of sad male')
# Short term Fourier Transform
X1 = librosa.stft(x1) #short time fourier trasnform : for time to frequency domain

# Amplitude of given frequency at a given time of an audio signal
Xdb1 = librosa.amplitude_to_db(abs(X1))
librosa.display.specshow(Xdb1, sr=sr1, x_axis='time', y_axis='log')
plt.show()

plt.figure(figsize=(20, 5))
plt.title('Specshow of happy female')
# Short term Fourier Transform
X2 = librosa.stft(x2)

# Amplitude of given frequency at a given time of an audio signal
Xdb2 = librosa.amplitude_to_db(abs(X2))
librosa.display.specshow(Xdb2, sr=sr2, x_axis='time', y_axis='log')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
plt.title('Specshow of sad female')
# Short term Fourier Transform
X3 = librosa.stft(x3) #short time fourier trasnform : for time to frequency domain

# Amplitude of given frequency at a given time of an audio signal
Xdb3 = librosa.amplitude_to_db(abs(X3))
librosa.display.specshow(Xdb3, sr=sr3, x_axis='time', y_axis='log')
plt.show()

plt.figure(figsize=(20, 5))
plt.title('Specshow of happy male')
# Short term Fourier Transform
X4 = librosa.stft(x4)

# Amplitude of given frequency at a given time of an audio signal
Xdb4 = librosa.amplitude_to_db(abs(X4))
librosa.display.specshow(Xdb4, sr=sr4, x_axis='time', y_axis='log')
plt.show()
```

# Zero Cross Rating

At what time how much signal is changing its direction. (positive to negative)

By examining the zero-crossing rate plot, one can gain insights into the noisiness or noisefullness of the audio.

```python
zero_crossings = librosa.zero_crossings(x1)
print("Sum of zero crossing ", zero_crossings.sum())
plt.figure(figsize=(20, 5))
plt.title('Zero Crossing Rate of sad male')
zcrs = librosa.feature.zero_crossing_rate(x1)
plt.plot(zcrs[0])
plt.show()

zero_crossings = librosa.zero_crossings(x2)
print("Sum of zero crossing ", zero_crossings.sum())
plt.figure(figsize=(20, 5))
plt.title('Zero Crossing Rate of happy female')
zcrs = librosa.feature.zero_crossing_rate(x2)
plt.plot(zcrs[0])
plt.show()
```

```python
#A zero crossing occurs when the sign of the audio signal changes from positive to negative or vice versa
#
zc = librosa.zero_crossings(x1)
print(sum(zc))
zc = librosa.zero_crossings(x2)
print(sum(zc))
```

# Spectral Centroids :
Where center of mass/sound is present

Wtg mean of Frequency

indicating where the "brightness" of the sound is located in the frequency domain

```python
from sklearn.preprocessing import minmax_scale
def normalize(x1, axis=0):
    return minmax_scale(x1, axis=axis)
```

```python
plt.figure(figsize=(20, 5))
plt.title('Spectral Centroids of sad male')
spectral_centroids = librosa.feature.spectral_centroid(y=x1, sr=sr1)[0]
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)
librosa.display.waveshow(x1,sr=sr1, alpha=0.4)
plt.plot(t, normalize(spectral_centroids), color='r')
plt.show()


#Red is frequency
```

# Spectral Rolloff

Frequency below specified perc of total spectral lies

It indicates how much energy is concentrated in the lower frequencies of the spectrum

```python
plt.figure(figsize=(20, 5))
plt.title('Spectral Rolloff of sad male')
spectral_rolloff = librosa.feature.spectral_rolloff(y=x1, sr=sr1, roll_percent=0.01)[0]
frames = range(len(spectral_rolloff))
t = librosa.frames_to_time(frames)
librosa.display.waveshow(x1,sr=sr1, alpha=0.4)
plt.plot(t, normalize(spectral_rolloff), color='r')
plt.show()
```

# MFCC and Mel Spectrogram

Gives Shape of Audio, What kind of properties audio poses

```python
plt.figure(figsize=(20, 6))
plt.title('MFCC Spectral of sad male')
mfccs = librosa.feature.mfcc(y=x1, sr=sr1)
print(mfccs.shape) #no of mfccs,no of frames
librosa.display.specshow(mfccs, sr=sr1, x_axis='time')
plt.show()

plt.figure(figsize=(20, 6))
plt.title('MFCC Spectral of happy male')
mfccs = librosa.feature.mfcc(y=x2, sr=sr2)
print(mfccs.shape)
librosa.display.specshow(mfccs, sr=sr2, x_axis='time')
plt.show()

#The colors in the spectrogram indicate the intensity or magnitude of each MFCC coefficient at different time frames.
#See red pard in fig
```

```python
plt.figure(figsize=(20, 6))
plt.title('Mel Spectrogram of sad male')
mel_spec = librosa.feature.melspectrogram(y=x1, sr=sr1)
librosa.display.specshow(mel_spec, sr=sr1, x_axis='time')
plt.show()

plt.figure(figsize=(20, 6))
plt.title('Mel Spectrogram of happy male')
mel_spec = librosa.feature.melspectrogram(y=x2, sr=sr2)
librosa.display.specshow(mel_spec, sr=sr2, x_axis='time')
plt.show()
```
